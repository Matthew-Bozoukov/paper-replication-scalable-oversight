```python
import json


with open("self_pref_indiv_gpt-4.1-nano.json", "r") as f:
    scores = json.load(f)
```

```python

import json
import matplotlib.pyplot as plt

model_map = {
    0: "Claude-3 Haiku",
    1: "Human",
    2: "Llama-2",
    3: "GPT-4",
    4: "GPT-3.5",
}

items = sorted(((int(k), v) for k, v in enumerate(scores)), key=lambda x: x[0])
indices, values = zip(*items)
labels = [ model_map[i] for i in indices ]  

plt.figure(figsize=(7,4))
bars = plt.bar(indices, values, width=0.6)
plt.xticks(indices, labels, rotation=45, ha="right")
plt.ylim(0,1)
plt.ylabel("Self-preference score")
plt.title("GPT-4.1-Nano individual preference")


for bar, val in zip(bars, values):
    plt.text(bar.get_x()+bar.get_width()/2, val+0.02, f"{val:.2f}",
             ha="center", va="bottom")

plt.tight_layout()
plt.show()
```
